```python
#IMPORT THIS HEADER
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.stats import linregress
import pylab
import math
from scipy.optimize import curve_fit
import random as ran
%matplotlib inline
```

```python
TDC=np.loadtxt(open("ShuweiMay5Fe.txt"))
```

```python
TDCs=[]

#filter
for i in range(len(TDC)):
    counter=0
    for n in range(len(TDC[0])):
        if TDC[i, n]!=-1:
            counter+=1           
    if counter>=3:
        TDCs.append(TDC[i,:])
```

```python
TDCs=np.asarray(TDCs)
```

```python
TDC8prefilter=TDCs[:, 7]*4/1000
TDC7prefilter=TDCs[:, 6]*4/1000
TDC6prefilter=TDCs[:, 5]*4/1000
TDC5prefilter=TDCs[:, 4]*4/1000
TDC4prefilter=TDCs[:, 3]*4/1000
TDC3prefilter=TDCs[:, 2]*4/1000
TDC2prefilter=TDCs[:, 1]*4/1000
TDC1prefilter=TDCs[:, 0]*4/1000
```

```python
TDC8prefilterlist=np.ndarray.tolist(TDC8prefilter)
TDC7prefilterlist=np.ndarray.tolist(TDC7prefilter)
TDC6prefilterlist=np.ndarray.tolist(TDC6prefilter)
TDC5prefilterlist=np.ndarray.tolist(TDC5prefilter)
TDC4prefilterlist=np.ndarray.tolist(TDC4prefilter)
TDC3prefilterlist=np.ndarray.tolist(TDC3prefilter)
TDC2prefilterlist=np.ndarray.tolist(TDC2prefilter)
TDC1prefilterlist=np.ndarray.tolist(TDC1prefilter)
```

```python
def remove_values_from_list(the_list, val):
    return [value for value in the_list if value != val]
```

```python
filteredTDC8array=np.sort(np.asarray(remove_values_from_list(TDC8prefilterlist, -0.004)))[::-1]
filteredTDC7array=np.sort(np.asarray(remove_values_from_list(TDC7prefilterlist, -0.004)))[::-1]
filteredTDC6array=np.sort(np.asarray(remove_values_from_list(TDC6prefilterlist, -0.004)))[::-1]
filteredTDC5array=np.sort(np.asarray(remove_values_from_list(TDC5prefilterlist, -0.004)))[::-1]
filteredTDC4array=np.sort(np.asarray(remove_values_from_list(TDC4prefilterlist, -0.004)))[::-1]
filteredTDC3array=np.sort(np.asarray(remove_values_from_list(TDC3prefilterlist, -0.004)))[::-1]
filteredTDC2array=np.sort(np.asarray(remove_values_from_list(TDC2prefilterlist, -0.004)))[::-1]
filteredTDC1array=np.sort(np.asarray(remove_values_from_list(TDC1prefilterlist, -0.004)))[::-1]

print(filteredTDC4array)
#TDC1 and TDC2 are empty, devoid of events
```

```python
plt.rcParams['figure.figsize']=16,7
data = filteredTDC3array
fig, ax = plt.subplots()
ax.hist(data, bins=500, range=(0, 16), facecolor=(0,0.6,0), edgecolor='black', histtype = 'step')

plt.yscale('log', nonposy='clip')
plt.xticks(range(0, 17))

plt.xlabel("Microsecond Bins", fontsize="15")
plt.ylabel("Counts", fontsize="15")
plt.title("TDC3 Raw Data: Iron Target", fontsize="18")

fig.savefig('Iron_R3.png') 
```
